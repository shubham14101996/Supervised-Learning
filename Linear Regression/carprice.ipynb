```python
cp.num_of_doors = cp.num_of_doors.replace({'four':4,'two':2})
```

```python
cp.num_of_cylinders = cp.num_of_cylinders.replace({'four':4,'six':6,'five':5,'eight':8,'two':2,'three':3,'twelve':12})
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cp[cp.select_dtypes(include=['object']).columns] = cp[cp.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
cp.head()
```

```python
from sklearn.model_selection import train_test_split
cp_train,cp_test = train_test_split(cp, test_size = 0.2, random_state = 100)
```

```python
cp_train_x = cp_train.iloc[:,0:-1]
cp_train_y = cp_train.iloc[:,-1]
```

```python
cp_test_x = cp_test.iloc[:,0:-1]
cp_test_y = cp_test.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
reg_cp = LinearRegression()
```

```python
reg_cp.fit(cp_train_x, cp_train_y)
```

```python
pred_train = reg_cp.predict(cp_train_x)
pred_test = reg_cp.predict(cp_test_x)
```

```python
Rsquare = reg_cp.score(cp_train_x, cp_train_y)
Rsquare
```

```python
N = cp_train_x.shape[0]
K = cp_train_x.shape[1]

Adj_Rsquare = 1-(1-Rsquare) * (N-1)/(N-K-1)
Adj_Rsquare
```
