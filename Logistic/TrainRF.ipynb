# Train Test Model 

```python
from sklearn.model_selection import train_test_split
df_train , df_test = train_test_split(df, test_size = .2)
```

```python
df_train_x = df_train.iloc[:,0:-1]
df_train_y = df_train.iloc[:,-1]

df_test_x = df_test.iloc[:,0:-1]
df_test_y = df_test.iloc[:,-1]
```

# Model 1

```python
from sklearn.tree import DecisionTreeClassifier
dt_Normal = DecisionTreeClassifier()
```

```python
dt_Normal.fit(df_train_x, df_train_y)
```

```python
pred_dt_Normal = dt_Normal.predict(df_test_x)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
tab_dt_Normal = confusion_matrix(pred_dt_Normal, df_test_y)
```

```python
tab_dt_Normal
```

```python
tab_dt_Normal[1]
```

```python
Acc_dt_Normal = tab_dt_Normal.diagonal().sum() * 100/ tab_dt_Normal.sum()
print('Accuracy of Normal Decision tree with all default values i.e criterion="gini index"',Acc_dt_Normal,'%')
```

```python
Error_class_zero = (tab_dt_Normal[0][0]/tab_dt_Normal[0].sum()*100).round(2)
print('Error for class Zero of First Confusion Matrix',Error_class_zero*100,'%')

Error_class_one = (tab_dt_Normal[1][1]/tab_dt_Normal[1].sum()*100).round(2)
print('Error for class One of First Confusion Matrix',Error_class_one,'%')

Error_class_two = (tab_dt_Normal[2][2]/tab_dt_Normal[2].sum()*100).round(2)
print('Error for class Two of First Confusion Matrix',Error_class_two*100,'%')

Error_class_three = (tab_dt_Normal[1][1]/tab_dt_Normal[1].sum()*100).round(2)
print('Error for class Three of First Confusion Matrix',Error_class_three,'%')
```

```python
dt_Normal.feature_importances_
```

```python
dt_Normal.feature_importances_.sum()
```

```python
feature_importance_df_Normal = pd.DataFrame({'Importance':dt_Normal.feature_importances_*100, 'variables':df_train_x.columns})
```

```python
feature_importance_df_Normal.sort_values('Importance', ascending= False).reset_index()
```

# Model 2 using Class Weight = 'Balanced'

```python
dt_balanced = DecisionTreeClassifier(max_depth = 6, min_samples_split = 150, class_weight='balanced')
print(dt_balanced.fit(df_train_x, df_train_y))
dt_balanced_pred = dt_balanced.predict(df_test_x)

from sklearn.metrics import confusion_matrix
tab_dt_balanced = confusion_matrix(dt_balanced_pred, df_test_y)
print(tab_dt_balanced)
print(dt_balanced.feature_importances_)
feature_importance_balanced = pd.DataFrame({'Importance':dt_balanced.feature_importances_*100, 'variables':df_train_x.columns})
feature_importance_balanced.sort_values('Importance', ascending=False)
```

# Model 3 Using Entropy as Criteria

```python
dt_balanced = DecisionTreeClassifier(criterion="entropy",max_depth = 6, min_samples_split = 150, class_weight='balanced')
print(dt_balanced.fit(df_train_x, df_train_y))
dt_balanced_pred = dt_balanced.predict(df_test_x)
print('---------------')
print('-----CONFUSION MATRIX------')
from sklearn.metrics import confusion_matrix
tab_dt_balanced = confusion_matrix(dt_balanced_pred, df_test_y)
print(tab_dt_balanced)
print('-----------------')
print('-----IMPORTANT FEATURES-------')
dt_balanced.feature_importances_
feature_importance_balanced = pd.DataFrame({'Importance':dt_balanced.feature_importances_*100, 'variables':df_train_x.columns})
feature_importance_balanced.sort_values('Importance', ascending=False)
```

# Model 4 Min Sample Split = 200

```python
dt_balanced = DecisionTreeClassifier(criterion="entropy",max_depth = 6, min_samples_split = 200, class_weight='balanced')
print(dt_balanced.fit(df_train_x, df_train_y))
dt_balanced_pred = dt_balanced.predict(df_test_x)
print('---------------')
print('-----CONFUSION MATRIX------')

from sklearn.metrics import confusion_matrix
tab_dt_balanced = confusion_matrix(dt_balanced_pred, df_test_y)
print(tab_dt_balanced)
print('-----------------')

print('-----IMPORTANT FEATURES-------')

dt_balanced.feature_importances_
feature_importance_balanced = pd.DataFrame({'Importance':dt_balanced.feature_importances_*100, 'variables':df_train_x.columns})
feature_importance_balanced.sort_values('Importance', ascending=False)
```
