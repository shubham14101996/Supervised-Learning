```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
Att.OverTime.value_counts()
```

```python
Att.Attrition = Att.Attrition.replace({'No':0, 'Yes':1})
Att.BusinessTravel = Att.BusinessTravel.replace({'Travel_Rarely':0, 'Travel_Frequently':1, 'Non-Travel':2})
Att.Department = le.fit_transform(Att.Department)
Att.EducationField = le.fit_transform(Att.EducationField)
Att.Gender = le.fit_transform(Att.Gender)
Att.JobRole = le.fit_transform(Att.JobRole)
Att.MaritalStatus = le.fit_transform(Att.MaritalStatus)
Att.OverTime = le.fit_transform(Att.OverTime)
```

```python
Att.head()
```

```python
Att.info()
```

```python
Att1 = Att.drop('Over18', axis = 1)
```

```python
Att1.columns
```

```python
Att[['Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department',
       'DistanceFromHome', 'Education', 'EducationField', 'EmployeeCount',
       'EmployeeNumber', 'EnvironmentSatisfaction', 'Gender', 'HourlyRate',
       'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction',
       'MaritalStatus', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
       'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel',
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']]
```

```python
Att1.Attrition.value_counts()
```

```python
from sklearn.model_selection import train_test_split
Att_train, Att_test = train_test_split(Att1, test_size = 0.2)
```

```python
Att_train_x = Att_train[['Age', 'BusinessTravel', 'DailyRate', 'Department',
       'DistanceFromHome', 'Education', 'EducationField', 'EmployeeCount',
       'EmployeeNumber', 'EnvironmentSatisfaction', 'Gender', 'HourlyRate',
       'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction',
       'MaritalStatus', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
       'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel',
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']]


Att_train_y = Att_train.iloc[:,1]


Att_test_x = Att_test[['Age', 'BusinessTravel', 'DailyRate', 'Department',
       'DistanceFromHome', 'Education', 'EducationField', 'EmployeeCount',
       'EmployeeNumber', 'EnvironmentSatisfaction', 'Gender', 'HourlyRate',
       'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction',
       'MaritalStatus', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
       'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel',
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']]


Att_test_y = Att_test.iloc[:,1]
```

```python
Att_train_y.value_counts()
```

```python
print(Att_train_x.shape)
print(Att_train_y.shape)
print(Att_test_x.shape)
print(Att_test_y.shape)
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
logreg.fit(Att_train_x, Att_train_y)
```

```python
pred_test = logreg.predict(Att_test_x)
pred_test
```

```python
from sklearn.metrics import confusion_matrix
tab_log = confusion_matrix(pred_test, Att_test_y)
tab_log
```
