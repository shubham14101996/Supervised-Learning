```python
from sklearn.model_selection import train_test_split
emmi_train, emmi_test = train_test_split(emmi, test_size = 0.2)
```

```python
emmi_train_x = emmi_train.drop(columns=['y'])
emmi_train_y = emmi_train.iloc[:,0]

emmi_test_x = emmi_test.drop(columns=['y'])
emmi_test_y = emmi_test.iloc[:,0]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

lin_emm = LinearRegression()
```

```python
lin_emm.fit(emmi_train_x , emmi_train_y)
pred_emm_lin_test = lin_emm.predict(emmi_test_x)
mse_lin_emm_test = np.mean(np.square(emmi_test_y - pred_emm_lin_test))

mape_lin_emm_test = np.mean(np.abs(emmi_test_y - pred_emm_lin_test) * 100/emmi_test_y )

# --- below for train
pred_emm_lin_train = lin_emm.predict(emmi_train_x)
mse_lin_emm_train = np.mean(np.square(emmi_train_y - pred_emm_lin_train))
mape_lin_emm_train = np.mean(np.abs(emmi_train_y - pred_emm_lin_train) * 100/emmi_train_y )
```
